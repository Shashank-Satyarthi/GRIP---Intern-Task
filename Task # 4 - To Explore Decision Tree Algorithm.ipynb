```python
#Loading the Datasets

import pandas as pd
df = pd.read_csv('Iris.csv')
df.head()
```

```python
input = df.drop(['Species','Id'],axis=1)
input.head()
```

```python
target = df.Species
```

```python
from sklearn.datasets import load_iris
iris_dataset = load_iris()
```

```python
iris_dataset.keys()
```

### Defining Decision Tree Algorithm

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
```

```python
model.fit(input,target)
```

### Visualization of Decision Tree

```python
from sklearn import tree
import matplotlib.pyplot as plt
%matplotlib inline
import graphviz
import seaborn as sns
import numpy as np
```

```python
# Import necessary libraries for graph viz
from sklearn.externals.six import StringIO  
from IPython.display import Image  
from sklearn.tree import export_graphviz
import pydotplus

# Visualize the graph
dot_data = StringIO()
export_graphviz(model, out_file=dot_data, feature_names=iris_dataset.feature_names,  
                filled=True, rounded=True,
                special_characters=True)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
Image(graph.create_png())
```

## Another approach

```python
Tree = tree.export_graphviz(model,out_file= None,feature_names = iris_dataset.feature_names,class_names = iris_dataset.target_names,filled = True,rounded=True,
                special_characters=True)
graph = graphviz.Source(Tree)
graph
```
